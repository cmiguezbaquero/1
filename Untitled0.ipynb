```python
import pandas as pd
```

```python
data = {'Producto' : ['Manzanas', 'Peras', 'Naranajas', 'Manzanas'],
        'Cantidad' : [10, 15, 8, 12],
        'Precio' : [1.5, 2.0, 1.2, 1.5]
        }
```

```python
df_cdv = pd.read_csv('./data/datos.csv')
df_excel = pd.read_excel('./data/datos.xlsx')
print (df_cdv)
```

```python
df = pd.DataFrame(data)
print (df)
```

```python
print(df.head (2))
```

```python
filtro = df [df['Cantidad'] > 10]
print (filtro)
```

```python
df ['Total'] = df ['Cantidad'] * df ['Precio']
print (df)
```

```python
print (df.describe())
```

```python
df_ordenado = df.sort_values (by = 'Cantidad', ascending= False)
print (df_ordenado)
```

```python
df_renombrado = df.rename (columns = {'Producto': 'Fruta', 'Cantidad': 'Unidades'})
print(df_renombrado)
```

```python
print (df['Producto'].unique ())
```

```python
print(df['Producto'].nunique())
```

```python
print(df['Producto'].value_counts())
```

```python
df_con_nulos =df.copy ()
df_con_nulos.loc[1,'Precio'] = None
print(df_con_nulos)
```

```python
print(df_con_nulos.isnull().sum())
```

```python
df_con_nulos ['Precio'] = df_con_nulos ['Precio'].fillna(df_con_nulos ['Precio'].mean())
```

```python
print(df_con_nulos)
```

```python
print(f"Filas antes: {len(df)}")
df_sin_dup = df.drop_duplicates(subset=['Producto'])
print(f"Filas después: {len(df_sin_dup)}")
print(df_sin_dup)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8, 6))
plt.bar(df['Producto'], df['Cantidad'], color = ['red', 'green', 'orange', 'red'])
plt.title('Cantidad por producto')
plt.xlabel('Producto')
plt.ylabel('Cantidad')
plt.show()
```

```python
ventas = df.groupby('Producto')['Total'].sum()
plt.figure(figsize=(8, 6))
plt.pie(ventas, labels=ventas.index, autopct='%1.1f%%', colors= ['red', 'orange', 'green'])
plt.title('Porcentaje de ventas por producto')
plt.show()
```
